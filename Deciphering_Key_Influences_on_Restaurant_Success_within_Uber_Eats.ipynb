## Removing Unwanted Characters
The following snippets of codes facilitate the removal of unwanted characters from each column. Initially, we eliminated the empty spaces and the 'USD' currency symbol from the price column. This step enables us to create an aggregate function that computes the mean price of items per restaurant. Consequently, we removed the unwanted characters in the name column of the restaurants since all the restaurants that contained an '&' symbol also included a ';'. Then, we eliminated the presence of duplicate restaurants with varying prices, which depicted the cost of their distinct menu offerings.

```python
# Removing the USD from the price of item column
restaurants['price_of_item'] = restaurants['price_of_item'].str.strip(' USD')
```

```python
# Converting the prices into a float type
restaurants['price_of_item'] = restaurants['price_of_item'].astype(float)
# Grouping by restaurant id to find the mean prices for each restaurant
mean_prices = restaurants.groupby('id')['price_of_item'].mean().reset_index()
mean_prices['price_of_item'] = mean_prices['price_of_item'].round(2)
# Merging mean prices back into dataframe
restaurants = pd.merge(restaurants, mean_prices, on ='id', suffixes = ('','_mean'))
# Renaming the the price of item column to mean price per item
restaurants.rename(columns = {'mean_price_of_item':'price_of_item'}, inplace = True)
# Dropping the price_of_item column
restaurants.drop('price_of_item', axis = 1, inplace = True)
restaurants.head(3)
```

```python
def extract_state(address):
    # Regular expression pattern to extract state abbreviations
    state_pattern = r'\b([A-Z]{2})\b'
    match = re.search(state_pattern, address)   #Obtains two capital letters placed next to eachother on the address column
    if match:
        state = match.group(1)
        # Check if the extracted state is in the mapping
        if state in state_to_region:
            return state
    return None

# Function to map states to regions
def map_state_to_region(state):
    return state_to_region.get(state, 'other_region')  # Returns 'other_region' if state is not found
```

**Cleaning price column**:

A new column named 'price_category' was introduced to categorize restaurants based on their price ranges, given the prevalence of missing values in the original 'price_range' column. The price range was calculated using 'price_of_item_mean,' excluding outliers and setting the lower bound as the minimum value and the upper bound 1.5 times the interquartile range above the upper quartile. Restaurants with prices falling within the first 1/3 of the price range were classified as 'cheap_prices,' those within the last 1/3 as 'high-end_prices,' and the remainder as 'average_prices.' This method allows for more precise restaurant categorization based on average item prices.

```python
# Creating a categorical column based on the mean price of item per restaurant (category labels = cheap_prices, average_prices, high-end_prices)
# Check for the Interquartile range
iqr = (restaurants['price_of_item_mean'].quantile(q=0.75) - restaurants['price_of_item_mean'].quantile(q=0.25))*1.5
#Divide the price of item mean range into 3 portion for categorisations
#Sum up upper quatile and 1.5*interquatile range to exclude the extreme value and create upper bound.
#Minus the upper bound by the minimum value and divided it by 3 to divide the range into 3 equal portion.
interval  = (restaurants['price_of_item_mean'].quantile(q=0.75)+iqr-restaurants['price_of_item_mean'].min())/3
#low_p set the price at first 1/3 of the range
low_p = restaurants['price_of_item_mean'].min()+interval
#high_p is set the price at 2/3 of the range
high_p = low_p + interval
# Create a list of our conditions. This was based on items that were less than or equal low_p, between low_p and upper_p percentile, and greater than or equal to the high_p percentile
price_conditions = [
    (restaurants['price_of_item_mean'] <= low_p),
    (restaurants['price_of_item_mean'] > low_p) & (restaurants['price_of_item_mean'] < high_p),
    (restaurants['price_of_item_mean'] >=high_p)
    ]

# Create a list of the values we want to assign for each condition
price_values = ['cheap_prices', 'average_prices', 'high_end_prices']

# Create a new column and use np.select to assign values to it using our lists as arguments
restaurants['price_category'] = np.select(price_conditions, price_values)
```

# 10. Challenges and Limitation of our Dataset
_____________________________________________________________

We encountered significant challenges while handling the datasets, resulting in notable biases. These challenges primarily revolved around two aspects. Firstly, data cleaning presented a major hurdle. Given that the data was scraped from the web, approximately 45% of the entries in the restaurant file, encompassing scores and review counts, were missing. Imputing values for such a substantial number of missing entries was deemed inappropriate. Consequently, we made the decision to drop columns with high levels of missing data, such as the price range column from the original file. We then calculated a new price category to replace it and dropped rows with missing values. Fortunately, despite this approach resulting in some loss of information and analytical limitations, we still retained over half of the dataset. However, this process introduced biases into the remaining data, notably limited geographic coverage due to a paucity of data points in certain states and an overrepresentation of American-style restaurants in the dataset.

Another challenge during the data cleaning process was effectively transforming two columns, "full address" and "category," into usable data. Initially, our objective was to conduct regional analyses by classifying the state for each restaurant based on its zip code. However, inconsistencies between zip codes and the states listed in the full address necessitated a shift in our approach. We opted to use the "full address" column to extract state information, but this posed additional challenges due to non-standardized address formats. Fortunately, we were able to devise a method to extract valid state information from the addresses. Regarding the "category" column, discrepancies in descriptions for the same category of restaurants made standardization and classification challenging using raw data alone. Extensive online research enabled us to categorize restaurants by matching relevant descriptions into broader categories. However, this approach had limitations, particularly in accurately categorizing restaurants into specific cuisines. For instance, a steakhouse could belong to either American or another cuisine, but in our classification, it would be grouped as American. This likely contributed to the disproportionate representation of American restaurants in our dataset.
